```python
import numpy as np
import pandas as pd
```

```python
customers = pd.read_csv("data/olist_customers_dataset.csv")
geoloc = pd.read_csv("data/olist_geolocation_dataset.csv")
o_items = pd.read_csv("data/olist_order_items_dataset.csv")
o_payments = pd.read_csv("data/olist_order_payments_dataset.csv")
o_reviews = pd.read_csv("data/olist_order_reviews_dataset.csv")
orders = pd.read_csv("data/olist_orders_dataset.csv")
products = pd.read_csv("data/olist_products_dataset.csv")
sellers = pd.read_csv("data/olist_sellers_dataset.csv")
cat_name = pd.read_csv("data/product_category_name_translation.csv")
```

```python
geo = geoloc.groupby("geolocation_zip_code_prefix", as_index=False).agg({"geolocation_lng":"mean", "geolocation_lat":"mean"})
geo
```

```python
customers = pd.merge(customers, geo, left_on="customer_zip_code_prefix", right_on="geolocation_zip_code_prefix", how="inner")
customers.drop(["customer_zip_code_prefix", "geolocation_zip_code_prefix"], axis=1, inplace=True)
```

```python
customers.columns = ["customer_id","customer_unique_id","customer_city","customer_state", "customer_lng","customer_lat"]
customers
```

```python
orders.drop(["order_purchase_timestamp","order_approved_at","order_delivered_carrier_date","order_delivered_customer_date","order_estimated_delivery_date"], axis=1, inplace=True)
orders = pd.merge(orders, customers, on="customer_id", how="inner")
orders
```

```python
oo = orders.groupby(["customer_id"], as_index=False).agg({"order_id":"count"})
oo["order_id"].value_counts()
```

```python
o_items.drop(["order_item_id", "shipping_limit_date"], axis=1, inplace=True)
orders = pd.merge(orders, o_items, on="order_id", how="inner")
orders
```

```python
products = products[["product_id", "product_category_name"]]
orders = pd.merge(orders, products, on="product_id", how="inner")
orders
```

```python
sellers = pd.merge(sellers, geo, left_on="seller_zip_code_prefix", right_on="geolocation_zip_code_prefix", how="inner")
sellers.drop(["seller_zip_code_prefix", "geolocation_zip_code_prefix"], axis=1, inplace=True)
```

```python
sellers.columns = ["seller_id","seller_city","seller_state","seller_lng","seller_lat"]
sellers
```

```python
orders = pd.merge(orders, sellers, on="seller_id", how="inner")
orders
```

```python
orders.drop(["customer_unique_id"], axis=1, inplace=True)
orders
```

```python
orders.to_csv("clean_ds.csv")
```
